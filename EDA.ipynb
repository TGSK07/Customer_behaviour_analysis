```python
correct_mapping = {
    # Clothing
    "T-shirt":"Clothing", "Shirt": "Clothing", "Shorts": "Clothing",
    "Hoodie": "Clothing", "Pants": "Clothing", "Socks": "Clothing",
    "Jeans": "Clothing", "Blouse": "Clothing", "Skirt": "Clothing",
    "Sweater": "Clothing", "Dress": "Clothing",

    # Accessories
    "Sunglasses": "Accessories", "Gloves": "Accessories", "Jewelry": "Accessories", "Hat": "Accessories",
    "Handbag": "Accessories", "Backpack": "Accessories", "Belt": "Accessories", "Scarf": "Accessories", "Bag": "Accessories",

    # Electronics
    "Laptop": "Electronics", "Phone": "Electronics", "Headphones": "Electronics", "Watch": "Electronics",

    # Footwear
    "Shoes": "Footwear", "Sandals": "Footwear", "Sneakers": "Footwear", "Boots": "Footwear",

    # Outwear
    "Coat": "Outwear", "Jacket": "Outwear"
}
```

```python
clean_data['Category'] = clean_data['Item Purchased'].map(correct_mapping)
clean_data.groupby("Category")["Item Purchased"].unique()
```

```python
clean_data.groupby('Category')['Size'].unique()
```

```python
clean_data.groupby('Item Purchased')['Size'].unique()
```

### Treating Category

```python
# Treating null values or replacing all values in the Electronic Category
clean_data.loc[clean_data['Category']=='Electronics', 'Size']='Not Applicable'
```

```python
# Treating null values or replacing all values in the Accessories Category
clean_data.loc[clean_data['Category']=='Accessories', 'Size']='Free Size'
```

```python
# Treating null values or replacing all values in the Footwear Category
clean_data.loc[clean_data['Category']=='Footwear','Size']='Not Available'
```

```python
# Treating null values in the Clothing Category
clean_data.loc[(clean_data['Category']=='Clothing') & (clean_data['Size'].isnull()),'Size'] = clean_data.loc[data['Category']=='Clothing']['Size'].mode()[0]
```

```python
clean_data.groupby('Category')['Size'].unique()
```

### Treating Review Rating

```python
clean_data['Review Rating'] = clean_data['Review Rating'].fillna(clean_data.groupby('Item Purchased')['Review Rating'].transform('mean')).round(1)
```

### Treating Previous Purchased

```python
clean_data['Previous Purchases'] = clean_data['Previous Purchases'].fillna(0)
```

### Treating Purchase Amount (USD)

```python
clean_data['Purchase Amount (USD)'] = clean_data['Purchase Amount (USD)'].fillna(clean_data.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean')).round(2)
```

```python
clean_data.isnull().sum()
```

```python
# Remove Duplicates
clean_data = clean_data.drop_duplicates(subset=['Customer ID'], keep='first')
```

### Renaming the Headers

```python
clean_data.columns = clean_data.columns.str.replace(" ","_")
clean_data.columns = clean_data.columns.str.lower()
```

```python
clean_data = clean_data.rename(columns={"purchase_amount_(usd)": "purchase_amount_usd"})
```
